```python
import bs4
from bs4 import BeautifulSoup
import urllib
```

```python
with urllib.request.urlopen('https://pythonprogramming.net/parsememcparseface/') as file:
    soup = BeautifulSoup(file.read(), 'html.parser')

print(soup.title.string)
```

```python
for child in soup.html.descendants:
    if type(child) == bs4.element.Tag:
        print(child.name)
        print(child.attrs)
        print(child.string)
```

```python
nav = soup.find('nav')
for child in nav.descendants:
    if type(child) == bs4.element.Tag:
        print(child.name)
        print(child.attrs)
        print(child.string)
    
# getting specific value - support the content
specific_value = nav.find('a', href='/support/')
print(specific_value.string)
```

```python
paragraphs = soup.find_all('p')

for paragraph in paragraphs:
    print(paragraph.string)
```

```python
for url in soup.find_all('a'):
    url_string = url.get('href')
    if (url_string.startswith('http')):
        print(url_string)
```
