# Exercise 1

```python
import spacy
from string import punctuation
from spacy.lang.en.stop_words import STOP_WORDS

nlp=spacy.load('en_core_web_sm')

file_name = './data/spaceX_DP.txt'

with open(file_name, 'r') as reader:
    spacex_dp = reader.read()

spacex_dp_x = nlp(spacex_dp)
print(spacex_dp_x)
```

Word frequency and removing stopwords

```python
stop_words = list(STOP_WORDS)
spacex_dp_tokens = [token.text for token in spacex_dp_x]
freq_word={}
maximum_frequency = 0
for w in spacex_dp_tokens:
    w_lower = w.lower()

    if w_lower not in stop_words and w_lower.isalpha():
        if w_lower in freq_word.keys():
            freq_word[w_lower] += 1
        else:
            freq_word[w_lower] = 1
            
        if maximum_frequency < freq_word[w_lower]:
            maximum_frequency = freq_word[w_lower]
            
freq_word
```

Maximum frequency

```python
word_frequencies = {}

for word in freq_word.keys():
    word_frequencies[word] = (freq_word[word]/maximum_frequency)
    
word_frequencies
```

Sentence score

```python
sentence_score = {}

for sent in spacex_dp_x.sents:
    for word in sent:
        if word.text.lower() in freq_word.keys():
            if sent in sentence_score.keys():
                sentence_score[sent] += freq_word[word.text.lower()]
            else:
                sentence_score[sent] = freq_word[word.text.lower()]
                
sentence_score
```

Summarized sentences

```python
from heapq import nlargest

top30percent_sentence=int(0.3*len(sentence_score))
summarized_sentences = nlargest(top30percent_sentence, sentence_score)
print(summarized_sentences)
```

Summary string

```python
summary_text_sents = []

for sent in summarized_sentences:
    summary_text_sents.append(sent.text)

summary = ' '.join(summary_text_sents)
print(f'Summary length: {len(summary)}')
print(f'Original length: {len(spacex_dp)}')
print(summary)
```

Comparison with gensim summary

```python
from gensim.summarization import summarize

print ('Summary:')
print (summarize(spacex_dp, word_count=250))
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')
from textblob import TextBlob
```

```python
sid = SentimentIntensityAnalyzer()
polarity_score = sid.polarity_scores(spacex_dp)

testimonial = TextBlob(spacex_dp)
print(polarity_score)
print(testimonial.sentiment)
```

The VADER sentiment analysis categorizes most of the words as neutral (87%), but has a positive compound score, which means the general sentiment of the text is positive.

The TextBlob analysis categorizes the text as very neutral sentiment (neither positive or negative), while being slightly more objective in nature.
